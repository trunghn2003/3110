```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu giả cho chiều cao, trọng lượng và công việc
np.random.seed(42)
num_students = 2000

# Tạo dữ liệu chiều cao (m), trọng lượng (kg) và công việc (danh sách ngẫu nhiên)
heights = np.random.uniform(1.5, 2.0, num_students)  # Chiều cao giữa 1.5m và 2.0m
weights = np.random.uniform(50, 100, num_students)    # Trọng lượng giữa 50kg và 100kg
jobs = np.random.choice(['Job A', 'Job B', 'Job C', 'Job D'], num_students)  # Các công việc

# Tạo DataFrame
df = pd.DataFrame({
    'Height': heights,
    'Weight': weights,
    'Job': jobs
})

# Tạo các biểu đồ phân phối
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Biểu đồ phân phối chiều cao
axes[0].hist(df['Height'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Height (m)')
axes[0].set_xlabel('Height (m)')
axes[0].set_ylabel('Frequency')

# Biểu đồ phân phối trọng lượng
axes[1].hist(df['Weight'], bins=30, color='salmon', edgecolor='black')
axes[1].set_title('Distribution of Weight (kg)')
axes[1].set_xlabel('Weight (kg)')
axes[1].set_ylabel('Frequency')

# Biểu đồ phân phối số lượng công việc
job_counts = df['Job'].value_counts()
axes[2].bar(job_counts.index, job_counts.values, color='lightgreen', edgecolor='black')
axes[2].set_title('Distribution of Job Count')
axes[2].set_xlabel('Job')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
